1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales?
   - Data is broadly categorized into qualitative and quantitative data.
   - Qualitative Data (Categorical Data):
   - Qualitative data describes characteristics, attributes, or qualities that cannot be measured numerically. Instead, it is categorized based on labels or attributes.
   - Examples of Qualitative Data:

   - Eye color (blue, brown, green)
   - Marital status (single, married, divorced)
   - Types of cars (SUV, sedan, truck)
   - Types of Qualitative Data:
   - 1.Nominal Scale,2. Ordinal Scale.

   Quantitative Data (Numerical Data):
   - Quantitative data represents measurable quantities and can be expressed in numbers.

   - Examples of Quantitative Data:
   - Height of students (in cm or inches)
   - Number of books in a library
   - Temperature readings
   - Types of Quantitative Data:
   -  Interval Scale
   - Ratio Scale

   - NOMINAL SCALE:Data is categorized without any inherent order.
  - Categories are distinct but have no ranking.
   - Example: Blood type (A, B, AB, O), Gender (Male, Female, Other)

   - ORDINAL SCALE:Data is categorized with a meaningful order or ranking, but the difference between ranks is not uniform.
   - Example: Education level (High school, Bachelor's, Master's, Ph.D.), Customer satisfaction (Poor, Fair, Good, Excellent)

   - INTERVAL SCALE:Numeric values with equal intervals between them but no true zero point.
   - Example: Temperature in Celsius or Fahrenheit (0Â°C is not the absence of temperature), IQ scores

   - RATIO SCALE:Numeric values with equal intervals and a true zero point, meaning absence of the variable is possible.
   - Example: Weight (0 kg means no weight), Income (0 dollars means no money), Distance traveled (0 km means no movement)

    - Dispersion refers to the spread or variability of a dataset. It measures how much the data values deviate from the central tendency (mean, median, or mode). A higher dispersion means the data points are more spread out, while lower dispersion indicates they are closer together.
    - How Variance and Standard Deviation Measure the Spread of Data:

4. What is a box plot, and what can it tell you about the distribution of data?
   - A box plot is a graphical representation of a dataset's distribution. It displays central tendency, dispersion, and outliers, making it useful for understanding the spread and skewness of data.
   - What a Box Plot Can Tell You About the Distribution of Data:
   - Central Tendency (Median)
   - Spread (Variability)
   - Skewness (Symmetry or Asymmetry)
   - presence of outliers

11. Explain what a random variable is and differentiate between discrete and continuous random variables?
    - A random variable is a numerical outcome of a random experiment. It assigns numerical values to different possible outcomes of a probability experiment.

    - Discrete Random Variable:
    - A discrete random variable takes on a countable number of distinct values. These values are often whole numbers.

    -  Continuous Random Variable:
    - A continuous random variable takes on an infinite number of possible values within a given range. These values are measured rather than counted.
